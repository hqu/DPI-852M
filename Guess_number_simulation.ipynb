# Simulation of 2 players guessing a number between 1 and 1 million.

**prompt:** write python simulation: play 1 chooses a number between 1 and 1 million.  Player 2 guesse it in under 20 tries

```python
import random

def simulation():
    # Player 1: Choose a number between 1 and 1 million
    low = 1
    high = 1_000_000
    target = random.randint(low, high)

    print(f"Player 1 chosen number: {target}")

    attempts = 0

    # Player 2: Guess the number logic (Binary Search)
    # We maintain the current search range [current_low, current_high]
    current_low = low
    current_high = high

    while True:
        attempts += 1

        # Player 2 guesses the midpoint
        guess = (current_low + current_high) // 2

        print(f"Attempt {attempts}: Player 2 guesses {guess}", end="")

        if guess == target:
            print(" -> Correct!")
            break
        elif guess < target:
            print(" -> Too low")
            current_low = guess + 1
        else: # guess > target
            print(" -> Too high")
            current_high = guess - 1

    print(f"\nTotal attempts: {attempts}")
    if attempts <= 20:
        print("Result: SUCCESS (<= 20 attempts)")
    else:
        print("Result: FAILURE (> 20 attempts)")

if __name__ == "__main__":
    simulation()
```

```python
simulation()
```

```python
def roll_dice():
    roll_results = []
    for _ in range(2):
        roll = random.randint(1, 6)
        roll_results.append(roll)
    return roll_results

for i in range(10):
    print(roll_dice())
    roll_results_list = roll_dice()
    last_row = df.iloc[-1]
    roll_number = int(last_row["Time"])
    df.loc[len(df.index)] = [roll_number+1, roll_results_list[0], roll_results_list[1], roll_results_list[0] + roll_results_list[1]]
```

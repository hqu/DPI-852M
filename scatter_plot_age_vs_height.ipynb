# Visualize the relationship between age and height

## Step 1: Import the dataset Howell1 into a pandas dataframe
https://github.com/rmcelreath/rethinking/blob/master/data/Howell1.csv

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rmcelreath/rethinking/master/data/Howell1.csv', sep=';')
```

```python
df
```

## Step 2: Extract the list of ages and heights from the dataframe into 2 lists (age_list and height_list)

```python
age_list = df["age"].tolist()
height_list = df["height"].tolist()
```

```python
print(age_list)
```

```python
print(height_list)
```

## Step 3: Graph the scatterplot

```python
plt.scatter(age_list, height_list)
plt.show()
```

## Step 4: Filter out everyone over the age of 18

```python
filtered_df_children = df[df['age'] < 18] 
```

```python
age_list_filtered = filtered_df_children["age"].tolist()
height_list_filtered = filtered_df_children["height"].tolist()
```

```python
plt.scatter(age_list_filtered, height_list_filtered)
plt.show()
```

```python
col = filtered_df_children.male.map({0:'orange', 1:'blue'})
filtered_df_children.plot.scatter(x='age', y='height', c=col)
```

```python
filtered_df_adult = df[df['age'] >= 18]
age_list_filtered_adult = filtered_df_adult["age"].tolist()
height_list_filtered_adult = filtered_df_adult["height"].tolist()
```

```python
plt.scatter(age_list_filtered_adult, height_list_filtered_adult)
plt.show()
```

```python
col = filtered_df_adult.male.map({0:'orange', 1:'blue'})
filtered_df_adult.plot.scatter(x='age', y='height', c=col)
plt.legend(["male", "female"])
```
